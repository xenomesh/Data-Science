```python
# To calculate future population based on the current growth rate
print('Welcome to the world population calculator')

# To get current population
current_pop = int(input('Enter the current world population: '))

# To get current growth rate
growth_rate = float(input('Enter the current growth rate: (e.g, 1.14% would be 0.014): '))

print(f'{"Year":4}{"Estimated Population":>22}{"Change":>13}')

# To display data from year 1 to year 100
for year in range(1,201):
    future_pop = current_pop * (1+growth_rate)
    print(f'{year:>4d}{int(future_pop):>22d}', end='')
    print(f'{int(future_pop - current_pop):13d}')
    current_pop = int(future_pop)
```

```python
# Import statistics to find mean, median, mode, & sorted
import statistics as stats
grades = [85, 93, 45, 89, 85]
average = sum(grades) /len(grades)
print(f'Average ={average: >6.2f}')

print(f'Grades = {sorted(grades)}')
print(f'Mean = {stats.mean(grades): >6.2f}')
print(f'Median = {stats.median(grades): >6.2f}')
print(f'Mode = {stats.mode(grades): >6.2f}')
```
